```python
import torch.nn as nn
from sklearn.model_selection import StratifiedKFold, train_test_split, StratifiedShuffleSplit

import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
import torch.optim as optim
from Transformer_classify_model import TransformerClassifier
from coral_pytorch.losses import corn_loss
from multi_class_dataloader import TimeMultiClassDataset, collate_time_series    
import copy
from torch.nn import CrossEntropyLoss
from coral_pytorch.dataset import corn_label_from_logits
from helper import validate_model, plot_and_save_confusion_matrix, classification_metrics, train_model

pooling_method = 'attention'

label_df = pd.read_csv('multi_4_class.csv', index_col = 0)
label_df["class"] = label_df["class"]-1 # labels should start at 0

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
X = label_df.drop(columns=['class'])
y = label_df['class']
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), start=1):
    print (fold)
    train_df = label_df.iloc[train_idx].reset_index(drop=True)
    val_df   = label_df.iloc[val_idx].reset_index(drop=True)
    print(f"Fold {fold}: train={len(train_df)}, val={len(val_df)}")

    train_loader = DataLoader(
        TimeMultiClassDataset(train_df),
        batch_size=32, shuffle=True, collate_fn=collate_time_series
    )
    val_loader = DataLoader(
        TimeMultiClassDataset(val_df),
        batch_size=32, shuffle=False, collate_fn=collate_time_series
    )
    class_counts = train_df['class'].value_counts().values
    
    model = train_model(train_loader, val_loader, class_counts, device, pooling_method='pooling_method')
    label_list, pred_list, proba_list = validate_model(model, val_loader, device)

    cm_df = plot_and_save_confusion_matrix(
        y_true=label_list,
        y_pred=pred_list,
        labels=[0,1,2,3],
        display_labels=['1','2','3','4'],
        normalize=None,
        cmap='Blues',
        title="Multi-Class Confusion Matrix",
        save_path="confusion_matrix.png"  # or omit to skip saving
    )

    # Assume you have lists or arrays: y_true, y_pred, and y_proba
    metrics, report = classification_metrics(
        y_true=label_list,
        y_pred=pred_list,
        y_proba=proba_list,       # optional, omit for only discrete metrics
        class_labels=[0,1,2,3],
        top_k=2
    )
    print(report)
    # If you need the raw numbers:
    print(metrics['accuracy'], metrics['f1_macro'], metrics['roc_auc'])
```
